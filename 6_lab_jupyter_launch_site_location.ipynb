```python
# TASK 3: Calculate the distances between a launch site to its proximities

def calculate_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    coords_in_radians = [radians(lat1), radians(lon1), radians(lat2), radians(lon2)]
    lat1, lon1, lat2, lon2 = coords_in_radians

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    radius_earth_km = 6371  # Radius of Earth in kilometers
    distance = radius_earth_km * c

    return distance

# Create a new DataFrame to calculate distances to the nearest cities or points of interest
proximities = pd.DataFrame({
    'Name': ['City A', 'City B', 'City C'],
    'Lat': [28.4, 29.0, 27.9],
    'Long': [-80.6, -95.1, -81.2]
})

# Add distances to each launch site
for index, site in launch_sites_df.iterrows():
    site_lat, site_long = site['Lat'], site['Long']
    distances = proximities.apply(lambda row: calculate_distance(site_lat, site_long, row['Lat'], row['Long']), axis=1)
    proximities[f"Distance_to_{site['Launch Site']}"] = distances

proximities
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
